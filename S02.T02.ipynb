```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 1 Exercici 1                 ##
##########################################################################
# Exercici 1: Crea un np.array d'una dimensió, que inclogui l'almenys 8 
# nombres sencers, data type int64. 
#     1-Mostra la dimensió i 
#     2-la forma de la matriu. .

#import library
import numpy as np
## create Variables
myArray = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype='i8')

#Print Result (1)
print ("La dimensió de l'Array és: ", myArray.ndim, "D")
print ("La forma de la matriu és:" , myArray.shape)
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 1 Exercici 2                 ##
##########################################################################
# Exercici 2: De la matriu de l'exercici 1, calcula el valor mitjà dels 
# valors introduïts i resta la mitjana resultant de cada un dels valors de la matriu.

#Print Result (1)
print ("La mitja Aritmètica és: ", myArray.mean())
print ("Resultat de cada element: ")
for x in range(8):    
        print(myArray.mean()-myArray[x])
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 1 Exercici 3                 ##
##########################################################################
# Exercici 3: Crea una matriu bidimensional amb una forma de 5 x 5. 
# Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

# create Variables
myArray1 = np.array([[11, 2, 3, 4, 5],
                     [10, 4, 36, 37, 38],
                     [20, 22, 15, 27, 18],
                     [30, 41, 26, 17, 28],
                     [40, 14, 11, 19, 34]
                   ], dtype='i8')

print ("El valor màxim de la Matriu és:", np.amax(myArray1)) 
myArray2 = np.amax(myArray1,axis=0)
print("Els valors màxims de l'Index 0 són:",myArray2)
print ("El valor màxim del Índex 0 és:", np.amax(myArray2,axis=0))
myArray2 = np.amax(myArray1,axis=1)
print("Els valors màxims de l'Index 1 són:",myArray2)
print ("El valor màxim del Índex 1 és:", np.amax(myArray2,axis=0))
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 2 Exercici 4                 ##
##########################################################################
# # Exercici 4:Mostreu-me amb exemples de diferents matrius, la regla fonamental
# de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les 
# seves dimensions coincideixen o si una de les matrius té una mida d'1".

# create Variables
myA = np.array([[11, 2, 3],
                [ 8, 7,13]])
myB = np.array( [ 1, 2, 1])  
myC = np.array([[11, 2, 3],
                [ 1, 1, 1],
                [ 8, 7,13]])
myE = np.array([[11, 2, 3],
                [ 1, 1, 1],
                [ 8, 7,13]])
    
  #Examples broadcasting
print("Exemple Arrays 2D*1D:", myA*myB)
print("Exemple Arrays 3D+1D:", myC*myB)
print("Exemple Arrays 3D/3D:", myC%myE)
print("Exemple Arrays 3D/3D:", myC+3)
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 2 Exercici 5                 ##
##########################################################################
# Exercici 5: Utilitza la Indexació per extreure els valors d'una columna 
# i una fila de la matriu. I suma els seus valors.

# create Variables
myArray5 = np.array([[11, 2, 3, 4, 5],
                     [10, 4, 36, 37, 38],
                     [ 0, 2,  1,  7,  1],
                     [30, 41, 26, 17, 28],
                     [40, 14, 11, 19, 34]
                   ], dtype='i8')

myAux = myArray5 [2]
print("Agafem la fila 3:",myAux,". Suma total :", myAux.sum())
myAux = myArray5 [:,1]
print("Agafem la columna 2:",myAux,". Suma total :", myAux.sum())
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 2 Exercici 6                 ##
##########################################################################
# Exercici 6: Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat,
# agafant cada element i comprovant si es divideix uniformement per quatre.
# Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

# create Variables
myArray5 = np.array([[11, 2, 3, 4, 5],
                     [10, 4, 36, 37, 38],
                     [ 0, 2,  1,  7,  1],
                     [30, 41, 26, 17, 28],
                     [40, 14, 11, 19, 34]
                   ], dtype='i8')

myMask=np.zeros((5,5))


myMask = np.ma.masked_where((myArray5%4!=0),myArray5)
print(myMask)
print(myMask.mask)
```

```python
##########################################################################
##           DATA SCIENCE: S02.T02: Nivell 2 Exercici 7                 ##
##########################################################################
# Exercici 7: A continuació, utilitzeu aquesta màscara per indexar a la 
# matriu de números original. Això fa que la matriu perdi la seva forma original, 
# reduint-la a una dimensió, però encara obteniu les dades que esteu cercant.

##Result: to extract the elements that we are searching
print((np.ma.compressed(myMask)))
```
